```python
x = data.drop(columns=['Class'], axis=1)
y = data['Class']
```

```python
from sklearn.preprocessing import StandardScaler # split and then scale to prevent data leakage --- note it
```

```python
sc = StandardScaler()
x_scaler = sc.fit_transform(x)
```

```python
x_scaler [-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaler, y,
test_size=0.20,
random_state=42,
stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("Training Accuracy :", model.score(x_train, y_train))
print("Testing Accuracy :", model.score(x_test, y_test))
```

```python
from sklearn.metrics import classification_report, f1_score
```

```python
print(classification_report(y_test, y_pred))
```

```python
print("F1 Score:",f1_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model1 = RandomForestClassifier()
model1.fit(x_train, y_train)
```

```python
y_pred = model1.predict(x_test)
```

```python
print("Training Accuracy :", model1.score(x_train, y_train))
print("Testing Accuracy :", model1.score(x_test, y_test))
```

```python
print(classification_report(y_test, y_pred))
```

```python
print("F1 Score:",f1_score(y_test, y_pred))
```

```python
from xgboost import XGBClassifier

model2 = XGBClassifier(n_jobs=-1)
model2.fit(x_train, y_train)
```

```python
y_pred = model2.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
print("F1 Score:",f1_score(y_test, y_pred))
```

# Better priorities recall and F1 score over accuracy and that’s what fraud analyst prefer.

In these cases of highly imbalanced datasets, consider reporting on an alternative metric to accuracy for performance. F1 score or AUC would be more illustrative for comparing classifier performance.

Given the imbalance here, accuracy is not especially informative of classifier performance. If I read your notebook right, a naive “always negative” classifier would achieve 99.8% accuracy.
